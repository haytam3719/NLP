```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/sample.csv')
df
```

```python
import pandas as pd
import re

# Supprimer les mentions d'utilisateurs Twitter
def remove_mentions(text):
    return re.sub(r'@[\w_]+', '', text)


# Appliquer la fonction de nettoyage à la colonne 'text'
df['text'] = df['text'].apply(remove_mentions)

print(df.head())

```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Téléchargez les ressources NLTK (si vous ne les avez pas déjà téléchargées)
nltk.download('punkt')
nltk.download('wordnet')


# Tokenisation
df['tokens'] = df['text'].apply(word_tokenize)

# Stemming
stemmer = PorterStemmer()
df['stemmed'] = df['tokens'].apply(lambda x: [stemmer.stem(word) for word in x])

# Lemmatisation
lemmatizer = WordNetLemmatizer()
df['lemmatized'] = df['tokens'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# Affichez le DataFrame résultant
print(df.head())
```

Supress Emojis and Links

```python
import pandas as pd
import re


# Function to remove emojis from text
def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emojis
                               u"\U0001F300-\U0001F5FF"  # Symbols & Pictographs
                               u"\U0001F680-\U0001F6FF"  # Transport & Map Symbols
                               u"\U0001F700-\U0001F77F"  # Alchemical Symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U0001F004-\U0001F0CF"  # Additional Emoticons
                               u"\U0001F170-\U0001F251"  # Enclosed Ideographic Supplement
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

# Function to remove web links from text
def remove_links(text):
    return re.sub(r'http\S+', '', text)

# Apply the functions to the 'text' column in the DataFrame
df['text'] = df['text'].apply(remove_emojis).apply(remove_links)

# Display the modified DataFrame
print(df)
```

```python
# Function to remove punctuation from tokenized text
def remove_punctuation_from_tokens(tokens):
    return ' '.join([re.sub(r'[^\w\s]', '', token) for token in tokens])

# Apply the function to the 'text' column in the DataFrame
df['text'] = df['text'].apply(lambda x: remove_punctuation_from_tokens(x.split()))

# Display the modified DataFrame
print(df)
```

```python
# Function to keep only words (remove everything else)
def keep_only_words(text):
    return re.sub(r'[^A-Za-z\s]', '', text)

# Apply the function to the 'text' column in the DataFrame
df['text'] = df['text'].apply(keep_only_words)
print(df)
```
